```python
x= df.drop(labels = "House price of unit area",axis = 1)
y= df[["House price of unit area"]]
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y ,test_size = 0.25)
(x_train.shape, y_train.shape,x_test.shape, y_test.shape)
```

### Model building

### Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

### Linear regression predictions

```python
y_predict_lr = model.predict(x_test)
residual = y_test-y_predict_lr
```

```python
residual
```

```python
r2 = r2_score(y_test,y_predict_lr)
r2
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
Vif = []
for i in range(x_train.shape[1]):
    Vif.append(variance_inflation_factor(x_train,i))
```

```python
pd.DataFrame({"Vif" : Vif},index = df.columns[0:7]).T
```

```python
sns.displot(residual,kind = "kde")
```

```python
# homoscedasticity
plt.scatter(y_predict_lr,residual)
```

```python
plt.plot(residual)
```

### Decission regressor

```python
from sklearn.tree import DecisionTreeRegressor 
from sklearn.metrics import r2_score
Decission_tree_regressor = DecisionTreeRegressor(max_depth=2)
Decission_tree_regressor.fit(x_train,y_train)
```

```python
y_predict_DecisionTree = Decission_tree_regressor.predict(x_test)
y_predict_DecisionTree
```

```python
r2 = r2_score(y_test,y_predict_DecisionTree)
r2
```

### RandomForestRegressor

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
RandomForest = RandomForestRegressor(n_estimators = 50)
RandomForest.fit(x_train, y_train)  
```

```python
Y_pred = RandomForest.predict(x_test)  # test the output by changing values
Y_pred
```

```python
r2 = r2_score(y_test,Y_pred)
r2
```

### support vector regressor

```python
from sklearn.svm import SVR
```

```python
regressior = SVR(kernel='linear',C=1)
regressior.fit(x,y)
```

```python
y_pred = regressior.predict(x_test)
```

### Ridge regression

```python
# import ridge regression from sklearn library 
from sklearn.linear_model import Ridge 

# Train the model 
ridgeR = Ridge(alpha = 0.1) 
ridgeR.fit(x_train, y_train) 
y_pred_ridge = ridgeR.predict(x_test) 
```

```python
r2 = r2_score(y_test,y_pred_ridge)
r2
```

### Lasso 

```python
from sklearn.linear_model import Lasso 

# Train the model 
lasso = Lasso(alpha = 1) 
lasso.fit(x_train, y_train) 
y_pred_lasso = lasso.predict(x_test)
```

```python
r2 = r2_score(y_test,y_pred_lasso)
r2
```

### ElasticNet

```python
# import model 
from sklearn.linear_model import ElasticNet 

# Train the model 
e_net = ElasticNet(alpha = 1,l1_ratio=0.2) 
e_net.fit(x_train, y_train) 

# calculate the prediction and mean square error 
y_pred_elastic = e_net.predict(x_test) 
r2 = r2_score(y_test,y_pred_elastic)
r2
```

```python
#XG Boost
import xgboost as xg
xgb_r = xg.XGBRegressor(objective ='reg:linear',
                  n_estimators = 10, seed = 123)
```

```python
xgb_r.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score

pred_xg = xgb_r.predict(x_test)
r2 = r2_score(y_test,pred_xg)
r2
```
